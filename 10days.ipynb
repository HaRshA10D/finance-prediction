```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import Flatten
from keras.layers.convolutional import Conv1D
from keras.layers.convolutional import MaxPooling1D
from keras.utils import np_utils
```

```python
data = pd.read_csv('dat.csv')
dates = list(data['date'])
datesuniq = list(set(dates))
final = []
for i in datesuniq:
    databydate = data[data['date'] == i]
    times = list(databydate['time']//10000)
    timesuniq = list(set(times))
    for j in timesuniq:
        databytime = databydate[databydate['time']//10000 == j]
        thistime = []
        thistime.append(databytime['open'].mean())
        thistime.append(databytime['high'].mean())
        thistime.append(databytime['low'].mean())
        thistime.append(databytime['close'].mean())
        print(thistime)
        final.append(thistime)
```

```python
finalopen=[]
finalclose=[]
finalhigh=[]
finallow=[]
for i in final:
    finalopen.append(i[0])
    finalhigh.append(i[1])
    finallow.append(i[2])
    finalclose.append(i[3])
```

```python
featopen = []
featlabelopen = []
featclose = []
featlabelclose = []
feathigh = []
featlabelhigh = []
featlow = []
featlabellow = []
win = 10
for i in range(0,len(finalopen)-win-7):
    featopen.append(finalopen[i:i+win])
    featlabelopen.append(finalopen[i+win+7])
    featclose.append(finalclose[i:i+win])
    featlabelclose.append(finalclose[i+win+7])
    feathigh.append(finalhigh[i:i+win])
    featlabelhigh.append(finalhigh[i+win+7])
    featlow.append(finallow[i:i+win])
    featlabellow.append(finallow[i+win+7])
```

```python
import numpy as np
featopen = np.array(featopen)
featclose = np.array(featclose)
feathigh = np.array(feathigh)
featlow = np.array(featlow)
featlabelopen = np.array(featlabelopen)
featlabelclose = np.array(featlabelclose)
featlabelhigh = np.array(featlabelhigh)
featlabellow = np.array(featlabellow)
```

np.savetxt("feat15.csv", featopen, delimiter=",")
np.savetxt("label15.csv", featlabelopen, delimiter=",")

```python
from numpy import newaxis
#featopen = featopen*100
#featlabelopen = featlabelopen*100
featopen = featopen[:,:,newaxis]
featopen.shape
```

```python
def cnnmodel():
    model = Sequential()
    model.add(Conv1D(input_shape = (win, 1), filters = 32, kernel_size = 3, activation = 'relu'))
    model.add(MaxPooling1D())
    model.add(Conv1D(filters = 16, kernel_size = 2, activation = 'relu'))
    model.add(Flatten())
    model.add(Dense(8, activation = 'relu'))
    model.add(Dense(1))
    model.compile(loss='mean_absolute_error', optimizer='sgd', metrics=['mean_squared_error', 'mean_absolute_error', 'mean_absolute_percentage_error'])
    return model
```

```python
thisModel = cnnmodel()
```

```python
thisModel.summary()
```

```python
from sklearn.cross_validation import train_test_split
x_train,x_test,y_train,y_test = train_test_split(featopen,featlabelopen,test_size=0.2,random_state=4)
```

```python
history = thisModel.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=50,batch_size=8)
```

```python
scores = thisModel.evaluate(x_test, y_test, verbose=0)
print(scores)
```
